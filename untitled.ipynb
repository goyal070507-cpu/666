```python
# ==========================================
# TRAIN NLLB (WITH 4-BIT QLORA)
# ==========================================
import torch
from datasets import load_from_disk
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer, Seq2SeqTrainingArguments, Seq2SeqTrainer, DataCollatorForSeq2Seq, BitsAndBytesConfig
from peft import get_peft_model, LoraConfig, TaskType, prepare_model_for_kbit_training

print("Loading Dataset from Drive...")
# 1. Reload the dataset you saved earlier
raw_datasets = load_from_disk("bho_hin_dataset_saved")

# 2. Setup Tokenizer (Notice the specific Devanagari script tags)
nllb_model_name = "facebook/nllb-200-distilled-600M"
tokenizer_nllb = AutoTokenizer.from_pretrained(nllb_model_name, src_lang="bho_Deva")

def preprocess_nllb(examples):
    inputs = [ex["bho"] for ex in examples["translation"]]
    targets = [ex["hin"] for ex in examples["translation"]]
    # Tokenize the input and target texts
    return tokenizer_nllb(inputs, text_target=targets, max_length=128, truncation=True)

print("Tokenizing data...")
tokenized_nllb = raw_datasets.map(preprocess_nllb, batched=True)

print("Initializing 4-bit Quantization...")
# 3. Shrink the model footprint to 4-bit precision
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16
)

# Load the base model compressed
model_nllb = AutoModelForSeq2SeqLM.from_pretrained(
    nllb_model_name,
    quantization_config=bnb_config,
    device_map="auto" # Automatically uses the GPU efficiently
)

# 4. Apply LoRA (Train only 1% of the model)
model_nllb = prepare_model_for_kbit_training(model_nllb)
lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    target_modules=["q_proj", "v_proj"],
    lora_dropout=0.05,
    bias="none",
    task_type=TaskType.SEQ_2_SEQ_LM
)
model_nllb = get_peft_model(model_nllb, lora_config)
model_nllb.print_trainable_parameters()

# 5. Training Arguments (Optimized for Colab memory)
args_nllb = Seq2SeqTrainingArguments(
    output_dir="nllb_checkpoints",
    eval_strategy="epoch",          # ✅ renamed from evaluation_strategy
    learning_rate=2e-4,
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    gradient_accumulation_steps=4,
    weight_decay=0.01,
    save_total_limit=1,
    num_train_epochs=3,
    fp16=True,
    predict_with_generate=True,     # ✅ add this for seq2seq eval
)

trainer_nllb = Seq2SeqTrainer(
    model=model_nllb,
    args=args_nllb,
    train_dataset=tokenized_nllb["train"],
    eval_dataset=tokenized_nllb["test"],
    processing_class=tokenizer_nllb,   # ✅ renamed from tokenizer
    data_collator=DataCollatorForSeq2Seq(tokenizer_nllb, model=model_nllb),
)

print("\n🚀 Starting NLLB Training Process...")
trainer_nllb.train()

# 6. Save the final adapter model to your Drive
save_path = "nllb_bho_hin_final"
trainer_nllb.save_model(save_path)
print(f"✅ NLLB Training Complete! Model safely saved to: {save_path}")
```
