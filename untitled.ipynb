```python
# ==========================================
# TRAIN NLLB (WITH 4-BIT QLORA)
# ==========================================
import torch
from datasets import load_from_disk
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer, Seq2SeqTrainingArguments, Seq2SeqTrainer, DataCollatorForSeq2Seq, BitsAndBytesConfig
from peft import get_peft_model, LoraConfig, TaskType, prepare_model_for_kbit_training

print("Loading Dataset from Drive...")
# 1. Reload the dataset you saved earlier
raw_datasets = load_from_disk("bho_hin_dataset_saved")

# 2. Setup Tokenizer (Notice the specific Devanagari script tags)
nllb_model_name = "facebook/nllb-200-distilled-600M"
tokenizer_nllb = AutoTokenizer.from_pretrained(nllb_model_name, src_lang="bho_Deva")

def preprocess_nllb(examples):
    inputs = [ex["bho"] for ex in examples["translation"]]
    targets = [ex["hin"] for ex in examples["translation"]]
    # Tokenize the input and target texts
    return tokenizer_nllb(inputs, text_target=targets, max_length=128, truncation=True)

print("Tokenizing data...")
tokenized_nllb = raw_datasets.map(preprocess_nllb, batched=True)

print("Initializing 4-bit Quantization...")
# 3. Shrink the model footprint to 4-bit precision
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16
)

# Load the base model compressed
model_nllb = AutoModelForSeq2SeqLM.from_pretrained(
    nllb_model_name,
    quantization_config=bnb_config,
    device_map="auto" # Automatically uses the GPU efficiently
)

# 4. Apply LoRA (Train only 1% of the model)
model_nllb = prepare_model_for_kbit_training(model_nllb)
lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    target_modules=["q_proj", "v_proj"],
    lora_dropout=0.05,
    bias="none",
    task_type=TaskType.SEQ_2_SEQ_LM
)
model_nllb = get_peft_model(model_nllb, lora_config)
model_nllb.print_trainable_parameters()

# 5. Training Arguments (Optimized for Colab memory)
args_nllb = Seq2SeqTrainingArguments(
    output_dir="nllb_checkpoints",
    eval_strategy="epoch",          # ✅ renamed from evaluation_strategy
    learning_rate=2e-4,
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    gradient_accumulation_steps=4,
    weight_decay=0.01,
    save_total_limit=1,
    num_train_epochs=3,
    fp16=True,
    predict_with_generate=True,     # ✅ add this for seq2seq eval
)

trainer_nllb = Seq2SeqTrainer(
    model=model_nllb,
    args=args_nllb,
    train_dataset=tokenized_nllb["train"],
    eval_dataset=tokenized_nllb["test"],
    processing_class=tokenizer_nllb,   # ✅ renamed from tokenizer
    data_collator=DataCollatorForSeq2Seq(tokenizer_nllb, model=model_nllb),
)

print("\n🚀 Starting NLLB Training Process...")
trainer_nllb.train()

# 6. Save the final adapter model to your Drive
save_path = "nllb_bho_hin_final"
trainer_nllb.save_model(save_path)
print(f"✅ NLLB Training Complete! Model safely saved to: {save_path}")
```

```python
# ==========================================
# CELL 4: EVALUATION (NLLB)
# ==========================================
import evaluate
import torch
from tqdm import tqdm
from datasets import load_dataset

# Reload dataset if not in memory
if 'raw_datasets' not in globals():
    raw_datasets = load_from_disk("./bho_hin_dataset_saved_v2")  # replace with your actual dataset name

from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "facebook/nllb-200-distilled-600M"

tokenizer_nllb = AutoTokenizer.from_pretrained(model_name)
model_nllb = AutoModelForSeq2SeqLM.from_pretrained(model_name)
# Load Metrics
bleu = evaluate.load("sacrebleu")
meteor = evaluate.load("meteor")
bertscore = evaluate.load("bertscore")
comet_metric = evaluate.load("comet")

# Take a subset of 200 sentences for evaluation (to save time/memory)
test_subset = raw_datasets["test"].select(range(min(200, len(raw_datasets["test"]))))
sources = [ex["translation"]["bho"] for ex in test_subset]
references = [ex["translation"]["hin"] for ex in test_subset]

print("Generating NLLB Predictions...")
model_nllb.eval()
predictions = []

# Set source language for NLLB tokenizer
tokenizer_nllb.src_lang = "bho_Deva"

# Device setup
device = "cuda" #if torch.cuda.is_available() else "cpu"
model_nllb.to("cuda")

# Generate translations batch by batch to save memory
batch_size = 8
for i in tqdm(range(0, len(sources), batch_size)):
    batch_src = sources[i:i+batch_size]
    inputs = tokenizer_nllb(batch_src, return_tensors="pt", padding=True, truncation=True).to(device)

    with torch.no_grad():
        generated_tokens = model_nllb.generate(
            **inputs,
            forced_bos_token_id=tokenizer_nllb.convert_tokens_to_ids("hin_Deva"),
            max_length=128
        )

    decoded = tokenizer_nllb.batch_decode(generated_tokens, skip_special_tokens=True)
    predictions.extend(decoded)

# Calculate Metrics
print("\nCalculating Metrics...")
bleu_score = bleu.compute(predictions=predictions, references=[[r] for r in references])['score']
meteor_score = meteor.compute(predictions=predictions, references=references)['meteor']
bs = bertscore.compute(predictions=predictions, references=references, lang="hi")
bert_score_f1 = sum(bs['f1']) / len(bs['f1'])
comet_score = comet_metric.compute(predictions=predictions, references=references, sources=sources)['mean_score']

print("\n--- FINAL NLLB METRICS ---")
print(f"BLEU: {bleu_score:.2f}")
print(f"METEOR: {meteor_score:.4f}")
print(f"BERTScore (F1): {bert_score_f1:.4f}")
print(f"COMET: {comet_score:.4f}")
```

```python
from datasets import Dataset, DatasetDict
import os

def load_parallel_files(hindi_dir, bhojpuri_dir):
    pairs = []
    for domain in os.listdir(hindi_dir):
        h_domain = os.path.join(hindi_dir, domain)
        b_domain = os.path.join(bhojpuri_dir, domain)
        if not os.path.isdir(h_domain) or not os.path.isdir(b_domain):
            continue
        h_files = sorted(os.listdir(h_domain))
        b_files = sorted(os.listdir(b_domain))
        for fname in h_files:
            if fname not in b_files:
                continue
            h_path = os.path.join(h_domain, fname)
            b_path = os.path.join(b_domain, fname)
            try:
                with open(h_path, encoding='utf-8', errors='ignore') as hf, \
                     open(b_path, encoding='utf-8', errors='ignore') as bf:
                    h_lines = [l.strip() for l in hf if l.strip()]
                    b_lines = [l.strip() for l in bf if l.strip()]
                for h, b in zip(h_lines, b_lines):
                    pairs.append({"translation": {"hin": h, "bho": b}})
            except Exception as e:
                print(f"Skipping {fname}: {e}")
    return pairs

base = "./Hindi-Bhojpuri-treebank-translation"
print("Loading pairs...")
all_pairs = load_parallel_files(f"{base}/Hindi", f"{base}/Bhojpuri")
print(f"Total pairs loaded: {len(all_pairs)}")

# Split 90/10
split = int(0.9 * len(all_pairs))
train_data = all_pairs[:split]
test_data  = all_pairs[split:]

raw_datasets = DatasetDict({
    "train": Dataset.from_list(train_data),
    "test":  Dataset.from_list(test_data),
})

print(raw_datasets)
print(raw_datasets["test"][0])

# Save correctly this time
raw_datasets.save_to_disk("./bho_hin_dataset_saved_v2")
print("Saved to ./bho_hin_dataset_saved_v2")
```

```python
# ==========================================
# EVALUATION: TEST YOUR TRAINED MODEL
# ==========================================
import torch
from datasets import load_from_disk
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer, BitsAndBytesConfig
from peft import PeftModel
import evaluate
from tqdm import tqdm

# 1. Mount Drive (just in case this is a fresh session)


print("Loading Metrics...")
# 2. Load Evaluation Metrics
bleu = evaluate.load("sacrebleu")
meteor = evaluate.load("meteor")
bertscore = evaluate.load("bertscore")
comet_metric = evaluate.load("comet")

print("Loading Dataset...")
# 3. Load the Saved Test Data
raw_datasets = load_from_disk("./bho_hin_dataset_saved_v2")
test_data = raw_datasets["test"]

# OPTIONAL: Evaluate on a subset of 500 sentences to save time. 
# If you want to evaluate the whole test set, change this to: test_subset = test_data
subset_size = min(500, len(test_data))
test_subset = test_data.select(range(subset_size))

sources = [ex["translation"]["bho"] for ex in test_subset]
references = [ex["translation"]["hin"] for ex in test_subset]

print("Loading Base Model & Tokenizer...")
# 4. Load Tokenizer and Base NLLB Model in 4-bit
base_model_name = "facebook/nllb-200-distilled-600M"
adapter_path = "./nllb_bho_hin_final"

tokenizer = AutoTokenizer.from_pretrained(base_model_name, src_lang="bho_Deva")

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16
)

base_model = AutoModelForSeq2SeqLM.from_pretrained(
    base_model_name, 
    quantization_config=bnb_config, 
    device_map="auto"
)

print("Attaching Your Trained Adapter...")
# 5. Merge your trained adapter weights with the base model
model = PeftModel.from_pretrained(base_model, adapter_path)
model.eval() # Put model in evaluation mode

print(f"\n🚀 Generating Translations for {len(sources)} sentences...")
predictions = []
batch_size = 8  # Keep batch size small to prevent Out of Memory during inference

# 6. Generate Translations (Batch by Batch)
# We use tqdm to show a progress bar
for i in tqdm(range(0, len(sources), batch_size)):
    batch_src = sources[i : i + batch_size]
    
    # Tokenize input
    inputs = tokenizer(batch_src, return_tensors="pt", padding=True, truncation=True, max_length=128).to("cuda")
    
    # Generate translation (force output to be Hindi Devanagari)
    with torch.no_grad():
        generated_tokens = model.generate(
            **inputs, 
            forced_bos_token_id=tokenizer.lang_code_to_id["hin_Deva"], 
            max_length=128,
            num_beams=4 # Use beam search for higher quality translations
        )
    
    # Decode tokens back into readable text
    decoded_preds = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
    predictions.extend(decoded_preds)

print("\n📊 Calculating Final Scores...")

# 7. Calculate all requested metrics
bleu_score = bleu.compute(predictions=predictions, references=[[r] for r in references])['score']
meteor_score = meteor.compute(predictions=predictions, references=references)['meteor']

# BERTScore returns a list of F1 scores, we take the average
bs = bertscore.compute(predictions=predictions, references=references, lang="hi")
bert_score_f1 = sum(bs['f1']) / len(bs['f1'])

# COMET requires source, hypothesis (prediction), and reference
comet_score = comet_metric.compute(predictions=predictions, references=references, sources=sources)['mean_score']

# 8. Print Results!
print("\n" + "="*40)
print("🏆 FINAL EVALUATION METRICS 🏆")
print("="*40)
print(f"BLEU Score:      {bleu_score:.2f} (Higher is better)")
print(f"METEOR Score:    {meteor_score:.4f} (Higher is better)")
print(f"BERTScore (F1):  {bert_score_f1:.4f} (Higher is better)")
print(f"COMET Score:     {comet_score:.4f} (Higher is better)")
print("="*40)

# Print a few examples to see it with your own eyes
print("\n👀 3 Translation Examples:")
for i in range(3):
    print(f"\nBhojpuri (Input):  {sources[i]}")
    print(f"Hindi (Actual):    {references[i]}")
    print(f"Hindi (Predicted): {predictions[i]}")
```
